```python
from roboflow import Roboflow
import os

rf = Roboflow(api_key="YOUR_API_KEY")
project = rf.workspace("roboflow-universe-projects").project("construction-site-safety")
dataset = project.version(27).download("yolov8")

print(f"✅Done! The data is here: {dataset.location}")
```

```python
from ultralytics import YOLO

# Download the ‘small’ model - it is the best for speed/accuracy balance
model = YOLO('yolov8s.pt') 

# Start of training
results = model.train(
    data=f"{dataset.location}/data.yaml", 
    epochs=50,      # enough for start
    imgsz=640,      
    batch=16,       
    device=0,       # clearly indicate to use GPU
    name='safety_v1_run'
)
```

```python
from ultralytics import YOLO
import os

model_path = '../runs/detect/safety_v1_run/weights/best.pt'

# Check if the file exists before loading to be sure
if os.path.exists(model_path):
    model = YOLO(model_path)
    print("✅ Model loaded successfully!")
    
    # Run inference
    results = model.predict(source='https://ultralytics.com/images/bus.jpg', save=True, conf=0.5)
    print(f"Result saved to: {results[0].save_dir}")
else:
    print(f"❌ File not found at: {os.path.abspath(model_path)}")
```
